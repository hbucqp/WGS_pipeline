```python
import os
import re
```

```python
seq_dict = {}
for file in os.listdir():
    if file.endswith('.fa'):
        print(file)
        with open(file, 'r') as f:
            for line in f.readlines():
                line = line.strip()
                if re.search('>(.*)', line):
                    sample = re.search('>(.*)', line).group(1)
                    print(sample)
                else:
                    if sample not in seq_dict.keys():
                        seq_dict[sample] = line
                    else:
                        seq_dict[sample] = seq_dict[sample] + line
```
